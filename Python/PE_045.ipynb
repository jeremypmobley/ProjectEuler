Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...  
Pentagonal	 	Pn=n(3n−1)/2	 	1, 5, 12, 22, 35, ...  
Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...  
It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.

```python
def calc_triangle_num(num:int) -> int:
    """ Function to calculate the value of the triangle number of given input number """
    return int((num*(num+1))/2)
```

```python
for i in range(1,10):
    print(i, calc_triangle_num(i))
```

```python
def calc_pent_num(num:int) -> int:
    """ Function to calculate the value of the pentagonal number of given input number """
    return int((num*(3*num-1))/2)
```

```python
for i in range(1,10):
    print(i, calc_pent_num(i))
```

```python
def calc_hex_num(num:int) -> int:
    """ Function to calculate the value of the hexagonal number of given input number """
    return int(num*(2*num-1))
```

```python
for i in range(1,10):
    print(i, calc_hex_num(i))
```

```python
def find_next_tph_num(t:int,p:int,h:int, verbose=False):
    "Function to find next number that is triangular, pentagonal, and hexagonal"
    all_equal = False
    while all_equal is False:
        t_val = calc_triangle_num(t)
        p_val = calc_pent_num(p)
        h_val = calc_hex_num(h)
        if t_val == p_val == h_val:
            all_equal = True
        elif t_val < p_val:
            t += 1
        elif p_val < h_val:
            p += 1
        else:
            h += 1
        if verbose:
            print(t,p,h)
    return t_val
```

```python
%%time
print(find_next_tph_num(286,1,2))
```
