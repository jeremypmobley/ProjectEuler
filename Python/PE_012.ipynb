The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1  
 3: 1,3  
 6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28  
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import math
import numpy as np
```

```python
def get_num_factors(num:int) -> list:
    """ Function to return the number of factors of a given number """
    num_factors = 0
    for i in range(1,math.ceil(np.sqrt(num))):
        if num % i == 0:
            num_factors += 2
    return num_factors
```

```python
print(get_num_factors(6))
print(get_num_factors(21))
print(get_num_factors(28))
```

```python
%%time

val, update_val = 1,1
max_num_divisors = 1
while max_num_divisors < 501:
    val_num_factors = get_num_factors(val)
    if val_num_factors > max_num_divisors:
        max_num_divisors = val_num_factors
    update_val +=1
    val += update_val

print(val-update_val)
```
